```python
#Importamos las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
```

```python
#Definimos los datos de entrada

x1 = np.array([1, 0, 3, 5, 8, 3, 5])
x2 = np.array([1, 2, 3, 7, 9, 8, 6])
D = np.array([0, 0, 1, 1, 1, 1, 1])

z = np.zeros(7)
w1 = np.random.rand(1)
w2 = np.random.rand(1)
b = 1
alpha = 0.1

# Creamos listas para guardar el error y las iteraciones
errors = []
iterations = []

# Perceptron calculation
for epoch in range(100):  # Number of iterations
    total_error = 0
    for i in range(len(x1)):
        z[i] = (x1[i] * w1 + x2[i] * w2) - b
        if z[i] > 0:
            y = 1
        else:
            y = 0
        error = D[i] - y
        total_error += abs(error)
        w1 += alpha * error * x1[i]
        w2 += alpha * error * x2[i]
        b += alpha * error
    errors.append(total_error)
    iterations.append(epoch + 1)

# Plotting error over iterations
plt.plot(iterations, errors, marker='o')
plt.xlabel('Iterations')
plt.ylabel('Total Error')
plt.title('Total Error vs. Iterations')
plt.grid(True)
plt.show()

# Plotting data points
plt.scatter(x1, x2, c=D, cmap='viridis', label='Data Points')

# Plotting decision boundary
x_decision = np.linspace(min(x1), max(x1), 100)
y_decision = (-w1 * x_decision + b) / w2
plt.plot(x_decision, y_decision, '-r', label='Decision Boundary')

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Perceptron Decision Boundary')
plt.legend()
plt.show()
```
